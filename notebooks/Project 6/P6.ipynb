```c++
Grid::Grid(){
    for(int col = 0; col < COLS; col++){
    	for(int row = 0; row < ROWS; row++){
            m_grid[row][col] = ' ';
        }
    }
}
```

```c++
void Grid::print(){
    for(int row = 0; row < ROWS; row++){
        for(int col = 0; col < COLS; col++){
            cout << m_grid[row][col];
        }
        cout << endl;
    }
}
```

```c++
void Grid::set(int x, int y, char c){
    if(x < ROWS && y < COLS){
        m_grid[x][y] = c;
    }
}
```

```c++
Shape::Shape(int x, int y){
    m_x = x;
    m_y = y;
}
```

```c++
Square::Square(int x, int y, int size) : Shape(x, y){
    m_x = x;
    m_y = y;
    m_size = size;
}
```

```c++
void Square::draw(Grid &grid){
    for(int startX = m_x, count = 0; startX < (m_x+m_size); startX++){
        for(int startY = m_y; startY < (m_y+m_size); startY++){
            if(count == 0 || count == m_size-1){
                grid.set(startX, startY, '*');
            }else{
                grid.set(startX, m_y, '*');
                grid.set(startX, m_y+m_size-1, '*');
            }
        }
        count++;
    }
}
```

```c++
void draw(Grid &grid){
    for(int col = 0; col < COLS; col++){
        grid.set(0, col, '*');
        grid.set(23, col, '*');
    }
    for(int row = 0; row < ROWS; row++){
        grid.set(row, 0, '*');
        grid.set(row, 59, '*');
    }
}
```
